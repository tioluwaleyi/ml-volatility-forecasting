```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
tickers = ["SPY", "AAPL", "MSFT", "^VIX"]
```

```python
data = yf.download(
    tickers,
    start="2015-01-01",
    end="2025-12-31",
    auto_adjust=True
)
```
